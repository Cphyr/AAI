# Challenge 01 - Cluster the Unknown

Take the **penguins** dataset (`seaborn.load_dataset("penguins")` or fetch from OpenML: 344 penguins, 4 numeric measurements, 3 species). **Drop the species column** - it is your hidden ground truth for final evaluation only.

Your objective is to recover the species structure **without ever using the labels during modeling**, and to defend every choice you make.

Requirements:
* Preprocess deliberately: handle missing values, decide on scaling, and justify (what happens to k-means if flipper length in mm and bill depth in mm are left unscaled next to body mass in grams?).
* Determine the number of clusters using at least two of: elbow, silhouette, GMM-BIC, dendrogram gap. Report what each says - they may disagree.
* Run k-means, GMM, DBSCAN (or HDBSCAN), and agglomerative clustering. For each: a 2D visualization (PCA for plotting is fine) and 1-2 sentences on whether its cluster definition fits this data.
* Only at the end: reveal the species labels and score each method with the Adjusted Rand Index (`sklearn.metrics.adjusted_rand_score`). Did the internal metrics predict the ARI ranking?

Constraints:
* Labels may appear only in the final evaluation cell.

Bonus: one species has a sex-based size difference. Can you find k=6 structure (species x sex)? Which algorithm surfaces it first?
