# Challenge 01 - Kernel Design

Generate the following synthetic dataset (this is your ground truth - pretend you don't know it): two classes in 2D, where class membership depends only on the **angle** of the point, `y = sign(sin(3 * theta))`, with radius uniform in [0.5, 2] and 10% label noise (1000 train / 1000 test points, fixed seed).

Your objective is to reach **at least 85% test accuracy with an SVM**, and to understand *why* your kernel works.

Requirements:
* Baseline: linear SVM and RBF SVM with default parameters - report both.
* Tune the RBF's (C, gamma) with a proper grid search + cross-validation; show the CV heatmap.
* Design a **custom kernel or feature map** that exploits the structure (hint: what happens if you hand-craft features like `theta = arctan2(y, x)` or use a kernel on the angle?). `SVC(kernel=callable)` accepts any function returning the Gram matrix - verify yours is symmetric PSD by checking the eigenvalues of the Gram matrix.
* Compare: number of support vectors and test accuracy for linear / tuned RBF / custom. Explain the differences in 3-4 sentences using margin and feature-space arguments.

Constraints:
* SVMs only (no trees, no nets). Feature engineering is allowed - that is the point.

Bonus: plot the decision boundary of all three models side by side, marking support vectors.
