# Challenge 01 - Tabular Ensembles

In this challenge you will build the strongest possible model on the **Adult (Census Income)** dataset: predict whether income exceeds $50K from 14 mixed numeric/categorical features (~48k rows). Load it with `sklearn.datasets.fetch_openml("adult", version=2)`.

Your objective is to **maximize ROC-AUC on a held-out test set** (use `train_test_split` with `random_state=0`, `test_size=0.2`), and to **report an OOB or cross-validated estimate that predicts your test score within 0.01**.

Requirements:
* Compare at least: a single decision tree, a random forest, and one gradient-boosting library (XGBoost / LightGBM / CatBoost).
* Handle the categorical features in at least two different ways (e.g. one-hot vs CatBoost native `cat_features` / target encoding) and compare.
* Use early stopping for the boosted model; report the selected number of trees.
* Produce a permutation-importance plot for your best model and write 3 sentences interpreting it.

Constraints:
* No neural networks, no test-set peeking during model selection.
* Total training time under 10 minutes on a laptop.

Bonus: bootstrap a 95% confidence interval for your test AUC (resample the *test set* predictions 2000 times) - is your margin over the runner-up model actually significant?
