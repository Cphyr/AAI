# Challenge 01 - Sensor Fusion

Simulate a vehicle driving a 2D path for 300 s (piecewise: straight segments + two turns; you design the true trajectory). It carries three sensors:

1. A **GPS-like** position sensor: unbiased, sigma = 15 m, one fix per second, but it **drops out completely for 60 s** in the middle (urban canyon).
2. A **speed sensor** (odometer): measures speed (not velocity components), sigma = 0.5 m/s, 10 Hz, with a **slowly drifting multiplicative bias** (starts at 1.00, drifts to 1.05).
3. A **heading sensor** (compass): sigma = 2 degrees, 10 Hz, with occasional glitches (1% of readings are off by up to 40 degrees).

Your objective is to reconstruct the trajectory with **position RMSE below 10 m**, including during the GPS outage.

Requirements:
* Implement the fusion filter yourself (NumPy). An EKF is natural here (speed/heading enter nonlinearly: dx = v*cos(psi)*dt) - linearize, or reformulate cleverly.
* Handle the different sensor rates correctly (predict at 10 Hz; update with whichever measurement arrives).
* Gate the compass glitches with an innovation test (chi-squared threshold) - show it firing.
* Estimate the odometer bias online by augmenting the state - plot your bias estimate against the true drift.
* Report position RMSE: overall, during the outage, and for a filter that naively trusts all sensors without gating. Plot all trajectories.

Constraints:
* No filtering libraries (`filterpy` etc.) - the filter must be your own code.

Bonus: implement an RTS smoother pass and report how much the outage-period error improves offline.
