# Challenge 01 - Flag Anomalous Tracks in Unlabeled Data

20,000 tracks. No labels. Somewhere between 0.5 % and 2 % of them are anomalous — you do not know
which, and you do not know the rate.

**The deliverable is not a detector. It is a defended evaluation protocol.** Specifically: given a
budget of **500 analyst labels**, produce an estimate of precision *and* recall at your chosen
operating threshold, **with a confidence interval**, and defend every choice in it.

**Honest baselines** (Chapter 0
[L02](../../chapter_00_problem_formulation/lessons/L02_baselines.md)) — all three are cheap, so
there is no excuse:

1. Isolation forest on ~15 kinematic features.
2. kNN distance in the same feature space.
3. Negative log-likelihood under a constant-velocity Kalman filter fitted on the whole corpus
   (Chapter 2 [L01](../../chapter_02_state_space_filtering/lessons/L01_kalman_filter.md)).

**Evaluation protocol** (Chapter 1
[L03](../../chapter_01_ts_foundations/lessons/L03_temporal_evaluation.md)): tracks carry a
timestamp. Fit the normal-behaviour model on the **first 70 % of days** and score the last 30 %.
No random splits — drift is part of the problem.

```python
import matplotlib.pyplot as plt
import numpy as np

RNG = np.random.default_rng(99)
plt.rcParams.update({"figure.figsize": (11, 3.2), "axes.grid": True, "grid.alpha": 0.3})

N_TRACKS = 20000
TRUE_RATE = RNG.uniform(0.005, 0.02)          # you are not told this


def make_corpus(n=N_TRACKS, rate=TRUE_RATE, seed=0):
    """Tracks as (day, features-generating params); returns raw position sequences."""
    rng = np.random.default_rng(seed)
    day = rng.integers(0, 100, n)
    is_anom = rng.random(n) < rate
    kind = np.where(is_anom, rng.choice(["loiter", "dash", "erratic"], n), "normal")
    out = []
    for i in range(n):
        T = int(rng.integers(120, 240))
        v = rng.normal(200, 20)
        w = rng.normal(0, 0.002, T)
        if kind[i] == "loiter":
            w = 0.04 * np.sign(np.sin(2 * np.pi * np.arange(T) / rng.integers(40, 70)))
            v *= 0.6
        elif kind[i] == "dash":
            v *= 1.9
        elif kind[i] == "erratic":
            w = rng.normal(0, 0.02, T)
        # slow drift over the corpus: later days are sampled more sparsely (a real-world nuisance)
        drop = 0.05 + 0.25 * day[i] / 100
        hdg = np.cumsum(w) + rng.uniform(0, 2 * np.pi)
        speed = v * (1 + 0.05 * rng.normal(0, 1, T))
        pos = np.cumsum(np.c_[speed * np.cos(hdg), speed * np.sin(hdg)], axis=0)
        pos += rng.normal(0, 30, pos.shape)
        keep = np.sort(rng.choice(T, max(20, int(T * (1 - drop))), replace=False))
        out.append((day[i], np.arange(T)[keep].astype(float), pos[keep]))
    return out, is_anom, kind


corpus, TRUTH, KIND = make_corpus()
print(f"{len(corpus)} tracks over {max(c[0] for c in corpus) + 1} days")
print("The truth vector exists ONLY so this challenge can be graded. Using it anywhere except in")
print("Part 5 (where you check your estimate against it) defeats the entire exercise.")
```

## Part 2 - Build the score

Feature extraction is Chapter 3 territory: use the smoothed state, not finite differences. Then
pick your framing — density/likelihood, reconstruction, or discriminative-with-few-labels.

One trap this corpus is built to spring: **the sampling rate degrades over time** (later days drop
more points). Any detector that sees gap statistics will happily learn "recent = anomalous". Check
your top alerts against `day` before you believe anything.

```python
def features(track):
    """Minimal starter feature set. Extend it - and justify each addition."""
    _, t, p = track
    dt = np.maximum(np.diff(t), 1e-6)
    v = np.diff(p, axis=0) / dt[:, None]
    sp = np.linalg.norm(v, axis=1)
    hdg = np.unwrap(np.arctan2(v[:, 1], v[:, 0]))
    w = np.diff(hdg) / dt[1:]
    path = np.linalg.norm(np.diff(p, axis=0), axis=1).sum()
    return np.array([
        np.median(sp), sp.std(), np.percentile(sp, 95),
        np.median(np.abs(w)), np.percentile(np.abs(w), 95),
        np.mean(np.abs(w) > 0.02), np.sum(np.diff(np.sign(w)) != 0) / len(w),
        np.linalg.norm(p[-1] - p[0]) / max(path, 1e-9),
        len(t), t[-1] - t[0], len(t) / max(t[-1] - t[0], 1e-9),
    ])


F = np.array([features(c) for c in corpus])
days = np.array([c[0] for c in corpus])
print("feature matrix:", F.shape)
```

## Part 3 - Design the label budget (the core of this challenge)

You have 500 labels. Chapter 5 L01 tells you why labelling the top 500 scores is the wrong move:
it estimates precision at the top and gives you **no information about recall**.

Design a **stratified sample with inverse-probability weighting**:

1. Partition the score range into strata. Justify the boundaries (why these, why this many?).
2. Allocate the 500 labels across strata. High-score strata hold the positives, so sample them
   heavily — but **no stratum may be sampled at zero**, or recall becomes unestimable.
3. Estimate, with Horvitz-Thompson weights $1/\pi_h$:
   $\widehat{TP}, \widehat{FP}, \widehat{FN}$ and hence precision and recall at your threshold.
4. Put an interval on both. Bootstrap within strata is the least painful route.

The oracle below is your analyst: it costs one label per call and refuses after 500.

```python
class LabelBudget:
    def __init__(self, truth, budget=500):
        self._truth, self.budget, self.used, self.seen = truth, budget, 0, {}

    def label(self, i):
        i = int(i)
        if i in self.seen:
            return self.seen[i]
        if self.used >= self.budget:
            raise RuntimeError(f"budget exhausted at {self.budget} labels - that is the point")
        self.used += 1
        self.seen[i] = bool(self._truth[i])
        return self.seen[i]


oracle = LabelBudget(TRUTH, budget=500)
# TODO: score every track, define strata, sample, call oracle.label(), estimate with IPW.
```
