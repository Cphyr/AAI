# Challenge 01 - Classify Trajectory Types

600 tracks from four behaviour classes. Classify them with **three method families** and explain
what the differences between them tell you about the representation.

**The honest baseline to beat** (Chapter 0
[L02](../../chapter_00_problem_formulation/lessons/L02_baselines.md)): **1NN-DTW** on the
resampled speed/turn-rate channels. Requirement: beat it with at least one of the other two
families — **or explain why you cannot**, with evidence. "Could not beat DTW, here is why" is a
full-credit answer; "my ROCKET got 91 %" with no DTW number is not.

**Evaluation protocol** (Chapter 1
[L03](../../chapter_01_ts_foundations/lessons/L03_temporal_evaluation.md)):
tracks are generated by *platforms*, and each platform contributes several tracks. **Split by
platform**, not by track — the `platform` column is provided for exactly this reason. Use 5-fold
grouped CV, tune inside the fold, report per-class F1 and a confusion matrix.

**Required:** the post-mortem in Part 5.

```python
import matplotlib.pyplot as plt
import numpy as np

RNG = np.random.default_rng(17)
plt.rcParams.update({"figure.figsize": (11, 3.2), "axes.grid": True, "grid.alpha": 0.3})

CLASSES = ["cruise", "maneuver", "hold", "survey"]


def make_track(cls, rng, n=None):
    """Return (t, enu positions) for one track. Lengths differ on purpose."""
    n = n or rng.integers(180, 320)
    v = rng.normal(210, 25)
    w = np.zeros(n)
    if cls == "cruise":
        w += rng.normal(0, 0.002, n)
    elif cls == "maneuver":                     # a few sharp turns
        for _ in range(rng.integers(2, 5)):
            a = rng.integers(0, n - 30)
            w[a:a + rng.integers(12, 30)] = rng.choice([-1, 1]) * rng.uniform(0.03, 0.07)
    elif cls == "hold":                         # sustained racetrack
        per = rng.integers(50, 80)
        w = 0.045 * np.sign(np.sin(2 * np.pi * np.arange(n) / per))
    elif cls == "survey":                       # lawnmower: straight legs + 180 turns
        leg = rng.integers(35, 55)
        for a in range(leg, n, leg + 12):
            w[a:a + 12] = np.pi / 12
    hdg = np.cumsum(w) + rng.uniform(0, 2 * np.pi)
    speed = np.full(n, v) * (1 + 0.05 * rng.normal(0, 1, n))
    pos = np.cumsum(np.c_[speed * np.cos(hdg), speed * np.sin(hdg)], axis=0)
    pos += rng.normal(0, 35, pos.shape)          # sensor noise
    keep = np.sort(rng.choice(n, int(n * rng.uniform(0.75, 1.0)), replace=False))
    return np.arange(n)[keep].astype(float), pos[keep]


tracks, labels, platforms = [], [], []
for pid in range(120):                           # 120 platforms, 5 tracks each
    rng = np.random.default_rng(1000 + pid)
    cls = CLASSES[pid % 4]
    for _ in range(5):
        tracks.append(make_track(cls, rng))
        labels.append(cls)
        platforms.append(pid)
labels = np.array(labels)
platforms = np.array(platforms)
print(f"{len(tracks)} tracks, {len(np.unique(platforms))} platforms, "
      f"lengths {min(len(t) for t, _ in tracks)}-{max(len(t) for t, _ in tracks)}")

fig, ax = plt.subplots(1, 4, figsize=(13, 3))
for a, c in zip(ax, CLASSES):
    i = int(np.where(labels == c)[0][0])
    p = tracks[i][1]
    a.plot(p[:, 0] / 1e3, p[:, 1] / 1e3, lw=1)
    a.set_title(c, fontsize=9); a.set_aspect("equal")
plt.tight_layout()
```

## Part 1 - Representation decisions (make them explicitly)

Before any classifier, decide and *justify*:

* Which channels? Raw positions are translation- and rotation-dependent — a classifier fed them
  will learn where the platform happened to be. Speed / turn rate / curvature are invariant
  (Chapter 3 [L02](../../chapter_03_trajectory_data/lessons/L02_frames_and_features.md)).
* How do you get turn rate from noisy, irregularly sampled positions? (Chapter 3
  [L01](../../chapter_03_trajectory_data/lessons/L01_irregular_sampling.md) — the answer is not
  finite differences.)
* Equal length or not? DTW tolerates ragged input; ROCKET does not.
* Do you z-normalise? Speed magnitude may be *the* discriminator between two classes — check
  before normalising it away.

## Part 2 - Family 1: DTW-1NN (the baseline)

Implement DTW with a Sakoe-Chiba band and tune the band width by inner CV. Cost is $O(nm)$ per
pair, so subsample if you need to — and say that you did.

```python
def dtw(a, b, band=None):
    """DTW distance between two (T, d) sequences, optional Sakoe-Chiba band."""
    n, m = len(a), len(b)
    band = band or max(n, m)
    D = np.full((n + 1, m + 1), np.inf)
    D[0, 0] = 0.0
    for i in range(1, n + 1):
        lo = max(1, int(i * m / n) - band)
        hi = min(m, int(i * m / n) + band)
        for j in range(lo, hi + 1):
            c = np.sum((a[i - 1] - b[j - 1]) ** 2)
            D[i, j] = c + min(D[i - 1, j], D[i, j - 1], D[i - 1, j - 1])
    return np.sqrt(D[n, m])


# TODO: build your channel representation, then 1NN-DTW under grouped 5-fold CV by platform.
```

## Part 4 - Family 3: ROCKET

Implement MiniROCKET-style random convolutional features: sample kernels with random dilation and
bias, convolve, extract **PPV** (proportion of positive values) and max, then ridge-classify. It is
about 30 lines — the starter below is most of them.

```python
def rocket_features(X, n_kernels=2000, seed=0):
    """X: (N, T, C) fixed-length, multi-channel. Returns (N, 2 * n_kernels) features."""
    rng = np.random.default_rng(seed)
    N, T, C = X.shape
    feats = np.zeros((N, 2 * n_kernels), dtype=np.float32)
    for k in range(n_kernels):
        length = rng.choice([7, 9, 11])
        w = rng.normal(0, 1, (length, C))
        w -= w.mean(0)
        dil = int(2 ** rng.uniform(0, np.log2((T - 1) / (length - 1))))
        bias = rng.uniform(-1, 1)
        idx = np.arange(length) * dil
        L = T - idx[-1]
        if L <= 0:
            continue
        conv = np.zeros((N, L))
        for j, off in enumerate(idx):
            conv += X[:, off:off + L, :] @ w[j]
        conv += bias
        feats[:, 2 * k] = (conv > 0).mean(1)          # PPV - the feature that matters
        feats[:, 2 * k + 1] = conv.max(1)
    return feats


# TODO: resample every track to a fixed length, build X (N, T, C), then ridge on rocket_features.
```
