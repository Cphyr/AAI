# Capstone - Reproduce a Central Claim

Two days. Pick a paper, reproduce **one** central claim on data it was not fitted to, and write up
where and why it does or does not replicate. This is the whole course applied to somebody else's
result.

Work through Chapter 9 [L01](../lessons/L01_critical_reading.md) first — the reading procedure
there is the first half of this project, and your **written prediction** (step 5 of that lesson)
must be committed before you run anything.

## 1 - Choose a target

Preferred targets, in rough order of how well they fit this course:

| Target | Claim shape | What you will re-implement |
| --- | --- | --- |
| **IMM vs. a learned trajectory predictor** | "the learned model beats the classical filter by X %" | Chapter 2 IMM + a small seq2seq or MDN (Ch.7 L03) |
| **A ROCKET bake-off claim** | "method M is not statistically distinguishable from HIVE-COTE" | Chapter 4: ROCKET, DTW-1NN, a feature pipeline |
| **A trajectory-prediction result** (ADE/FDE on an AV or ATC dataset) | "architecture A improves min-ADE by X" | Chapter 7 L03, plus a constant-velocity baseline |
| **A TS anomaly-detection result** | "detector D achieves F1 = X" | Chapter 5, including the point-adjustment trap |

Constraints that make this doable in two days: the claim must be **one number on one dataset**, the
method must be re-implementable in under ~200 lines or available as a package, and you must be able
to construct a **new or held-out dataset** for it (a different subset, a different region, a
different platform class, or a defensible simulator).

Get the choice approved before starting.

## 2 - Requirements

Non-negotiable, all four:

1. **Honest baselines** (Chapter 0
   [L02](../../chapter_00_problem_formulation/lessons/L02_baselines.md)). The naive one for the
   task — persistence, constant velocity, seasonal naive, DTW-1NN, majority class — tuned with the
   same effort you gave the method.
2. **Clean temporal evaluation** (Chapter 1
   [L03](../../chapter_01_ts_foundations/lessons/L03_temporal_evaluation.md)). Rolling-origin or
   grouped-blocked splits, preprocessing inside the fold, and a stated splitting unit. Include one
   leakage probe (label shuffle, or a one-step shift test).
3. **At least one "change-the-assumption" mini-experiment.** Pick one and do it properly:
   * noisier sensors (inflate measurement noise 2-5x),
   * a different sampling rate (halve it; watch for aliasing, Ch.1 L02),
   * more clutter or lower $P_D$ (Ch.2 L04),
   * a regime change or distribution shift in the test period (Ch.6),
   * a different geographic region or platform class (tests the invariance you assumed).
4. **A write-up** — see section 4.

And one requirement on process: **state your prediction before you run**. "I expect the gain to
shrink from 12 % to under 4 % once the baseline gets the same lookback" is a scientific claim; you
should be right about half the time, and being wrong is where you learn.

result, and it is worth more than a rushed number.

## 4 - The write-up (the deliverable)

Four to six pages, in this order:

1. **The claim.** One sentence with a number, dataset and metric, plus the paper's protocol as you
   reconstructed it (not as the abstract described it).
2. **Your protocol.** Diagram: what was trained on what, evaluated on what, split how, tuned where.
   Every deviation from the original, and why.
3. **Results.** Baselines, the method, fold-level distributions, and a statistical statement on the
   *difference* (Diebold-Mariano for forecasts, Wilcoxon over folds/datasets for classification).
   Say plainly whether the claim replicated, replicated partially, or did not.
4. **The assumption experiment.** What you changed, what happened, and what that says about which
   assumption the original result depended on.
5. **The assumption map.** The table from Ch.9 L01 step 7: every assumption from Chapters 1-7, does
   it hold in the original setting, does it hold in yours.
6. **Your prediction, scored.** What you wrote in step 1 of the schedule, and how it turned out.
7. **What you would do next**, with the effort estimate.

## 5 - How this is assessed

| Weight | Criterion |
| --- | --- |
| 30 % | Protocol quality: is the evaluation actually clean, and can someone else re-run it? |
| 25 % | Baselines: present, tuned, and honestly compared |
| 20 % | The assumption experiment and what you concluded from it |
| 15 % | Clarity and calibration of the claims — including saying "I do not know" where you do not |
| 10 % | Reproducibility of your own work: seeds, config, one command to re-run |

Note what is **not** on this list: whether the paper's claim held up. That is not under your
control and is not what is being graded.

## 6 - Starter harness

One place to put the protocol so that every method in your comparison is forced through the same
door. Extend it; the point is that the split logic lives in exactly one function.

```python
import numpy as np


def rolling_origin_splits(n, folds=5, train_frac=0.5, embargo=0):
    """Yield (train_idx, test_idx) with an embargo gap. The only place splitting is decided."""
    start = int(n * train_frac)
    step = (n - start) // folds
    for k in range(folds):
        tr_end = start + k * step
        te_start = tr_end + embargo
        te_end = min(te_start + step, n)
        if te_end <= te_start:
            break
        yield np.arange(tr_end), np.arange(te_start, te_end)


def run_comparison(methods, data, metric, **split_kw):
    """methods: {name: fn(train_idx, test_idx) -> predictions}. Returns per-fold scores."""
    results = {k: [] for k in methods}
    for tr, te in rolling_origin_splits(len(data), **split_kw):
        for name, fn in methods.items():
            results[name].append(metric(fn(tr, te), data[te]))
    return {k: np.array(v) for k, v in results.items()}


def diebold_mariano(loss_a, loss_b, h=1):
    """DM statistic on per-step losses; |DM| > 1.96 is significant at 5%."""
    d = np.asarray(loss_a) - np.asarray(loss_b)
    T = len(d)
    dbar = d.mean()
    gamma = [np.mean((d[k:] - dbar) * (d[:T - k] - dbar)) for k in range(h)]
    var = (gamma[0] + 2 * sum(gamma[1:])) / T
    return dbar / np.sqrt(max(var, 1e-12))


# sanity check on the harness itself, so you are not debugging it at midnight
rng = np.random.default_rng(0)
a, b = rng.normal(1.0, 0.3, 500), rng.normal(1.0, 0.3, 500)
print(f"DM on two identical-distribution loss streams: {diebold_mariano(a, b):+.2f} (expect |DM| < 2)")
print(f"DM when b is genuinely better:                 {diebold_mariano(a, b - 0.2):+.2f} (expect >> 2)")
print("folds:", [(len(tr), len(te)) for tr, te in rolling_origin_splits(1000, embargo=20)])
```
