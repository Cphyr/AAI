# Challenge 01 - Regime-Change Coverage

A series with three regime changes of different character: a **variance jump**, a **level shift**,
and a **dynamics change** (the autocorrelation structure itself changes). Build 90 % prediction
intervals with at least three methods, show which one maintains coverage and which fails silently,
**and propose plus implement a mitigation**.

**Honest baselines** (Chapter 0
[L02](../../chapter_00_problem_formulation/lessons/L02_baselines.md)):

1. A constant-width Gaussian interval from the training residual std.
2. Split conformal on the same point predictor.

**Evaluation protocol** (Chapter 1
[L03](../../chapter_01_ts_foundations/lessons/L03_temporal_evaluation.md)): strictly chronological
train / calibration / test blocks. Report **rolling coverage** (window 100) and interval width
together — never coverage alone, since $(-\infty,\infty)$ covers perfectly.

**Required:** the mitigation in Part 4 and the post-mortem in Part 5.

```python
import matplotlib.pyplot as plt
import numpy as np

RNG = np.random.default_rng(4242)
plt.rcParams.update({"figure.figsize": (11, 3.4), "axes.grid": True, "grid.alpha": 0.3})
ALPHA = 0.10

N = 4000
BREAKS = {900: "variance x3", 1900: "level shift +6", 2900: "dynamics: phi 0.4 -> 0.92"}
```

```python
t = np.arange(N)
y = np.zeros(N)
eps_scale = np.where(t < 900, 1.0, 3.0)
phi = np.where(t < 2900, 0.4, 0.92)
level = np.where(t < 1900, 0.0, 6.0)
season = 2.5 * np.sin(2 * np.pi * t / 24)
for i in range(1, N):
    y[i] = phi[i] * (y[i - 1] - level[i - 1] - season[i - 1]) + RNG.normal(0, eps_scale[i])
    y[i] += level[i] + season[i]

plt.plot(y, lw=0.5)
for b, lbl in BREAKS.items():
    plt.axvline(b, c="r", ls="--")
    plt.text(b, y.max(), " " + lbl, fontsize=7, va="top")
plt.title("Three regime changes, three different kinds of assumption failure")
plt.tight_layout()

L = 24
X = np.array([y[i - L:i] for i in range(L, N)])
X = np.c_[X, np.sin(2 * np.pi * t[L:] / 24), np.cos(2 * np.pi * t[L:] / 24)]
Y = y[L:]
i_tr, i_ca = 600, 800                                  # train and calibrate BEFORE the first break
print(f"train {i_tr}   calibration {i_ca - i_tr}   test {len(Y) - i_ca} "
      f"(all three breaks are in the test period, on purpose)")
```

## Part 1 - Build the intervals

At minimum:

* constant-width Gaussian (baseline 1),
* split conformal (baseline 2),
* **CQR** — conformalised quantile regression,
* **ACI** — adaptive conformal, $\alpha_{t+1}=\alpha_t+\eta(\alpha-\text{err}_t)$,

and optionally a Bayesian structural model whose filter widens $P_t$ by itself. The
[walkthrough](../walkthroughs/lesson_three_intervals.ipynb) has runnable versions of all of these
— reuse them rather than re-deriving.

```python
def rolling_coverage(inside, w=100):
    return np.convolve(inside.astype(float), np.ones(w) / w, "valid")


def report(name, y_true, lo, hi, breaks=(900 - L - i_ca, 1900 - L - i_ca, 2900 - L - i_ca)):
    ins = (y_true >= lo) & (y_true <= hi)
    edges = [0] + [b for b in breaks if b > 0] + [len(y_true)]
    seg = [ins[a:b].mean() for a, b in zip(edges[:-1], edges[1:])]
    print(f"{name:<26} overall {ins.mean():.3f}  width {np.mean(hi - lo):6.2f}  "
          + "  ".join(f"R{i}:{v:.3f}" for i, v in enumerate(seg)))
    return ins


# TODO: fit your point predictor and the four interval methods, then call report() for each.
```

## Part 4 - Mitigation (required)

Implement **at least one**, and evaluate it against the unmitigated version:

* **Regime-aware calibration** — run online change-point detection (Chapter 3
  [L03](../../chapter_03_trajectory_data/lessons/L03_segmentation.md)) on the residual stream and
  reset/refit the calibration set at each detected change. Report the detection delay and what
  coverage does *during* that delay — the delay is the cost, and it is unavoidable.
* **Dynamic nonconformity scores** — normalise residuals by a rolling volatility estimate,
  $s_t = |y_t-\hat\mu_t| / \hat\sigma_t$, so the score distribution is stable even when the
  residual scale is not.
* **Weighted / decayed calibration** — exponentially down-weight old residuals (NexCP-style), so
  the calibration set forgets the pre-break regime.
* **ACI with a tuned $\eta$**, or an aggregate over several $\eta$ (AgACI).

Report for the mitigated method: overall coverage, per-regime coverage, mean width, and the
**Winkler score** — a mitigation that fixes coverage by doubling the width has not fixed anything.
